```python
import pandas as pd
from sklearn.model_selection import train_test_split
from keras import models, layers, optimizers
from sklearn.preprocessing import MinMaxScaler

from keras.models import Sequential
from keras import models,layers
from keras.utils import to_categorical
```

```python
data_path = './datasets/haberman.data'
df = pd.read_csv(data_path, sep=",", header=None)
```

```python
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns, index=df.index)
```

```python
df
```

```python
y = df[3]  # Target column
X = df.drop(3, axis=1)  # Feature columns

# Splitting the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

(X_train, X_test, y_train, y_test)
```

```python
network = models.Sequential()
network.add(layers.Dense(100, activation='relu', input_shape=(X_train.shape[1],)))
network.add(layers.Dense(1, activation='sigmoid'))

# Configure the Adam optimizer with a custom learning rate
learning_rate = 0.10  # Specify your desired learning rate here
adam_optimizer = optimizers.Adam(learning_rate=learning_rate)

network.compile(optimizer=adam_optimizer, loss="binary_crossentropy", metrics=["accuracy"])
```

```python
X_train.shape[1]
```

```python
network.fit(X_train, y_train, epochs=100, batch_size=40)
```

```python
predicted = network.predict(X_test)
test_loss, test_acc = network.evaluate(X_test, y_test)
print(test_acc, test_loss)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
y_pred_prob = network.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype('int32')  # Convert probabilities to binary labels
test_loss, test_acc = network.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_acc}, Test Loss: {test_loss}")

# Generating the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```
