```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import preprocessing
```

```python
df = pd.read_csv('spam.csv')
df.head()
```

```python
df.describe()
```

```python
df[df['Category'] == 'ham'].value_counts()
```

```python
# Encoding the 'Category' column
label_encoder = preprocessing.LabelEncoder()
df['Category'] = label_encoder.fit_transform(df['Category'])
df
```

```python
vectorizer = CountVectorizer(stop_words='english', lowercase=True)
```

```python
x = vectorizer.fit_transform(df.Message)
y = df['Category']
x_train_vectorized, x_test_vectorized, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
nb = MultinomialNB()
nb.fit(x_train_vectorized, y_train)
```

```python
train_score = nb.score(x_train_vectorized, y_train)
test_score = nb.score(x_test_vectorized, y_test)
print(f"Training score: {train_score}")
print(f"Test score: {test_score}")
```

```python
testing_emails = ['hi, wanna hangout at 10? i heard there is a good chinese restaurant nearby. see you soon'
                 ,"don't miss this chance to win 100$ dollars"]
```

```python
testing_emails_vectorized = vectorizer.transform(testing_emails)
predictions = nb.predict(testing_emails_vectorized)
print(predictions)
```
